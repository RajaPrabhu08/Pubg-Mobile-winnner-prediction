# Feature Engineering

```python
data['totalDistance'] = data['rideDistance'] + data['swimDistance'] + data['walkDistance']
```

```python
data['totalHeals'] = data['boosts'] + data['heals']
```

```python
data.head()
```

```python
data.drop(columns=['Id','groupId','matchId'],inplace=True)
data.drop(columns=['rideDistance','swimDistance','walkDistance','boosts','heals'],inplace=True)
```

```python
data1 = data.copy()
```

```python
data1.head()
```

```python
features = data1.drop(columns=['winPlacePerc'])
```

```python
target = data1['winPlacePerc']
```

Encoding

```python
features = pd.get_dummies(data1,dtype='int')
```

```python
features
```

```python
features.shape
```

```python
target.shape
```

# Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
sc.fit(features)
```

```python
features = pd.DataFrame(sc.transform(features))
```

```python
features
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(features,target, test_size=0.3, random_state=42)
```

```python
print(x_train.shape,y_train.shape)
print(x_test.shape,y_test.shape)
```

# Model Selection 

## Linear Regression

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(x_train, y_train)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
pred = lr.predict(x_test)
```

```python
#RMSE- Root Mean Square Error
from sklearn.metrics import mean_squared_error
```

```python
rmse = np.sqrt(mean_squared_error(y_test,pred))
```

```python
print('The RMSE Of LR Is:',rmse)
```

## Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dtree = DecisionTreeRegressor()
dtree.fit(x_train,y_train)
```

```python
pred = dtree.predict(x_test)
```

```python
rmse = np.sqrt(mean_squared_error(y_test,pred))
```

```python
print('The RMSE Of Dtree:',rmse)
```

## Voting Regressor

```python
from sklearn.ensemble import VotingRegressor
```

```python
vr = VotingRegressor(estimators=[('lr',lr),('dt',dtree)])
```

```python
vr.fit(x_train,y_train)
```

```python
pred = vr.predict(x_test)
```

```python
rmse = np.sqrt(mean_squared_error(y_test,pred))
```

```python
print('The RMSE Of VotingRegressor:',rmse)
```

## Catboost Regressor

```python
import catboost as cb
```

```python
model = cb.CatBoostRegressor(iterations=100,loss_function='RMSE')
model.fit(x_train,y_train)
```

```python
prediction = model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
rmse = np.sqrt(mean_squared_error(y_test,prediction))
print("RMSE: {:.3f}".format(rmse))
```

```python
from sklearn.metrics import mean_absolute_error,r2_score
```

```python
# pred on test set
y_pred = model.predict(x_test)
```

```python
mae = mean_absolute_error(y_test,y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print(f"Mean Absolute Error (MAE):{mae:.4f}")
print(f"R2 score: {r2:.4f}")
print(f"Average R2:",r2.mean())
```

```python
plt.figure(figsize=(15,6))
plt.scatter(y_test, y_pred, alpha=0.3, color='teal')
plt.plot([0,1],[0,1], '--r') # perfect prediction line
plt.xlabel("Actual winPlacePerc")
plt.ylabel("Predicted winPlacePerc")
plt.title("Actual Vs Prediction winPlacePerc")
plt.grid(True)
plt.show()
```
